```python
#3 Find the mean number of all-inclusive hotels across all destinations
data["no.All-inclusive hotels"].mean()
```

```python
#4 Find the lowest scoring destination

lowestscore = data["Feedback score"] == data["Feedback score"].min()
print("The destination with lowest score is:")
data[lowestscore]
```

```python
#5 Find the highest scoring destination
highestscore = data["Feedback score"] == data["Feedback score"].max()
print("The destination with higest score is:")
data[highestscore]
```

```python
#6 Find all the destinations where there are more than 9 all-inclusive hotels
filter_allinclusive = data["no.All-inclusive hotels"] > 9
data[filter_allinclusive]

```
